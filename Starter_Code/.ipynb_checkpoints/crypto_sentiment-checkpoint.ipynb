## 1. Sentiment Analysis

Use the [newsapi](https://newsapi.org/) to pull the latest news articles for Bitcoin and Ethereum and create a DataFrame of sentiment scores for each coin.

Use descriptive statistics to answer the following questions:
1. Which coin had the highest mean positive score?
2. Which coin had the highest negative score?
3. Which coin had the highest positive score?

```python
# Initial imports
import os
import pandas as pd
from dotenv import load_dotenv
import nltk as nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

%matplotlib inline
```

```python
# Additional imports
from newsapi import NewsApiClient
load_dotenv()
```

```python
# Read your api key environment variable
api_key = os.getenv("news_api")
print(type(api_key))          # Use type to verify that the "news_api" key was correctly retrieved. It should be string ('str')
```

```python
# Create a newsapi client
newsapi = NewsApiClient(api_key=api_key)
```

```python
# Fetch the Bitcoin news articles
bitcoin_news_en = newsapi.get_everything(
    q="bitcoin",
    language="en",
    page_size=100,
    sort_by="relevancy"
)

# Show the total number of Bitcoin news articles
bitcoin_news_en["totalResults"]

# Print total news articles about Bitcoin
print(f"Total news articles about Bitcoin: {bitcoin_news_en['totalResults']}")
```

```python
# Fetch the Ethereum news articles
ethereum_news_en = newsapi.get_everything(
    q="ethereum",
    language="en",
    page_size=100,
    sort_by="relevancy"
)

# Show the total number of Ethereum news articles
ethereum_news_en["totalResults"]

# Print total news articles about Ethereum
print(f"Total news articles about Ethereum: {ethereum_news_en['totalResults']}")
```

```python
# Create the Bitcoin sentiment scores DataFrame
bitcoin_sentiments = []

for article in bitcoin_news_en["articles"]:
    try:
        text = article["content"]
        date = article["publishedAt"][:10]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        bitcoin_sentiments.append({
            "text": text,
            "date": date,
            "compound": compound,
            "positive": pos,
            "negative": neg,
            "neutral": neu
            
        })
        
    except AttributeError:
        pass
    
# Create DataFrame
bitcoin_df = pd.DataFrame(bitcoin_sentiments)

# Reorder DataFrame columns
cols = ["date", "compound", "negative", "neutral", "positive", "text"]
bitcoin_df = bitcoin_df[cols]

bitcoin_df.head()
```

```python
# Create the Ethereum sentiment scores DataFrame
ethereum_sentiments = []

for article in ethereum_news_en["articles"]:
    try:
        text = article["content"]
        date = article["publishedAt"][:10]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        ethereum_sentiments.append({
            "date": date,
            "compound": compound,
            "negative": neg,
            "neutral": neu,
            "positive": pos,
            "text": text
        })
        
    except AttributeError:
        pass
    
# Create DataFrame
ethereum_df = pd.DataFrame(ethereum_sentiments)

# Reorder DataFrame columns
cols = ["date", "compound", "negative", "neutral", "positive", "text"]
ethereum_df = ethereum_df[cols]

ethereum_df.head()
```

```python
# Describe the Bitcoin Sentiment
bitcoin_df.describe()
```

```python
# Describe the Ethereum Sentiment
ethereum_df.describe()
```

###   Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word.
2. Remove Punctuation.
3. Remove Stopwords.

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re
```

```python
# Code to download corpora
import nltk
nltk.download('reuters')
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
# Instantiate the lemmatizer
lemmatizer = WordNetLemmatizer()

# Create a list of stopwords
print(stopwords.words('english'))

# Expand the default stopwords list if necessary
```

```python
# Complete the tokenizer function
def tokenizer(text):
    #"""Tokenizes text."""
    sw = set(stopwords.words('english'))
    # Remove the punctuation from text
    regex = re.compile("[^a-zA-Z ]")
    re_clean = regex.sub('', text)
  
    # Create a tokenized list of the words
    words = word_tokenize(re_clean)
    
    # Lemmatize words into root words
    lem = [lemmatizer.lemmatize(word) for word in words]
   
    # Convert the words to lowercase
    tokens = [word.lower() for word in lem if word.lower() not in sw]
    
    # Remove the stop words
    
    
    return tokens
```

```python
# Testing that the tokenizer function created is working
print(set(tokenizer(text)))
```

```python
# Create a new tokens column for Bitcoin
bitcoin_tokens_col = []

for article in bitcoin_news_en["articles"]:
    try:
        text = article["content"]
        date = article["publishedAt"][:10]
        tokens = tokenizer(text)
        
        bitcoin_tokens_col.append({
            "date": date,
            "text": text,
            "tokens": tokens     # New column with tokenized text
        })
        
    except AttributeError:
        pass
            
    
# Create DataFrame
bitcoin_tokens_col_df = pd.DataFrame(bitcoin_tokens_col)

# Reorder DataFrame columns
cols = ["date", "text", "tokens"]

bitcoin_tokens_col_df = bitcoin_tokens_col_df[cols]
bitcoin_tokens_col_df.head()
```

```python
# Creating a new data frame for Bitcoin by adding the tokenized text column
bitcoin_tokenized_df = pd.merge(bitcoin_df, bitcoin_tokens_col_df, how='inner')
bitcoin_tokenized_df.head()
```

```python
# Create a new tokens column for Ethereum
ethereum_tokens_col = []

for article in ethereum_news_en["articles"]:
    try:
        text = article["content"]
        date = article["publishedAt"][:10]
        tokens = tokenizer(text)
        
        ethereum_tokens_col.append({
            "date": date,
            "text": text,
            "tokens": tokens     # New column with tokenized text
        })
        
    except AttributeError:
        pass
            
    
# Create DataFrame
ethereum_tokens_col_df = pd.DataFrame(ethereum_tokens_col)

# Reorder DataFrame columns
cols = ["date", "text", "tokens"]
ethereum_tokens_col_df = ethereum_tokens_col_df[cols]

ethereum_tokens_col_df.head()
```

```python
# Creating a new data frame for Ethereum by adding the tokenized text column
ethereum_tokenized_df = pd.merge(ethereum_df, ethereum_tokens_col_df, how='inner')
ethereum_tokenized_df.head()
```

### NGrams and Frequency Analysis

In this section you will look at the ngrams and word frequency for each coin. 

1. Use NLTK to produce the n-grams for N = 2. 
2. List the top 10 words for each coin. 

```python
from collections import Counter
from nltk import ngrams
```

```python
# Preparing the data for Bitcoin N-grams

# Create an empty string
bitcoin_text_combined = ''
# Loop through all the headlines and add them to 'text_combined' 
for i in bitcoin_news_en['articles']:
    bitcoin_text_combined += i['title'] + ' ' + i['content'] + ' ' # add a space after every headline and body, so the first and last words are not glued together
# Print the first 1000 characters to screen for inspection
print(bitcoin_text_combined[0:1000])
```

```python
# Generate the Bitcoin N-grams where N=2
def bigram_counter_btc(bitcoin_text_combined): 
    processed_btc = tokenizer(bitcoin_text_combined)
    bigrams_btc = ngrams(processed_btc, n=2)
    top_10_btc = dict(Counter(bigrams_btc).most_common(10))
    return pd.DataFrame(list(top_10_btc.items()), columns=['bigram_btc', 'count'])

# Top Ten Bitcoin N-grams where N=2
bigram_counter_btc(bitcoin_text_combined)
```

```python
# Generate the Ethereum N-grams where N=2
def bigram_counter_eth(ethereum_text_combined): 
    processed_eth = tokenizer(ethereum_text_combined)
    bigrams_eth = ngrams(processed_eth, n=2)
    top_10_eth = dict(Counter(bigrams_eth).most_common(10))
    return pd.DataFrame(list(top_10_eth.items()), columns=['bigram_eth', 'count'])

bigram_counter_eth(ethereum_text_combined)
```

### Word Clouds

In this section, you will generate word clouds for each coin to summarize the news for each coin

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = [20.0, 10.0]
```

```python
# Data preparation
def process_text(bitcoin_text_combined):
    sw = set(stopwords.words('english'))
    regex = re.compile("[^a-zA-Z ]")
    re_clean = regex.sub('', bitcoin_text_combined)
    words = word_tokenize(re_clean)
    lem = [lemmatizer.lemmatize(word) for word in words]
    output = [word.lower() for word in lem if word.lower() not in sw]
    return ' '.join(output)
```

```python
# Generate the Bitcoin word cloud

input_text_btc = process_text(bitcoin_text_combined)
wc_btc = WordCloud().generate(input_text_btc)
plt.imshow(wc_btc)
```

```python
# Generate the Ethereum word cloud
input_text_eth = process_text(ethereum_text_combined)
wc_eth = WordCloud().generate(input_text_eth)
plt.imshow(wc_eth)
```

## 3. Named Entity Recognition

In this section, you will build a named entity recognition model for both Bitcoin and Ethereum, then visualize the tags using SpaCy.

```python
import spacy
from spacy import displacy
```

```python
# Download the language model for SpaCy
# !python -m spacy download en_core_web_sm
```

```python
# Load the spaCy model
nlp = spacy.load('en_core_web_sm')
```

---
### Bitcoin NER

```python
# Concatenate all of the Bitcoin text together

# Create an empty string
bitcoin_all_text = ''
# Loop through all the headlines and add them to 'text_combined' 
for i in bitcoin_news_en['articles']:
    bitcoin_all_text += i['title'] + ' ' + i['content'] + ' ' # add a space after every headline and body, so the first and last words are not glued together
# Print the first 1000 characters to screen for inspection
print(bitcoin_all_text[0:1000])
```

```python
# Run the NER processor on all of the text
# Set article to be analyzed with spaCy
doc = nlp(bitcoin_all_text)

# Add a title to the document
doc.user_data["title"] = "Analysis of Bitcoin News Articles - Bitcoin NER"
```

```python
# Render the visualization
# Render NER visualization with displacy to determine entities for extraction
displacy.render(doc, style='ent')
```

```python
# List all Entities
# Extract geopolitical and organizational entities
entities = [ent.text for ent in doc.ents if ent.label_ in ['GPE', 'ORG', 'PERSON', 'NORP']]
```

```python
# lower and join each entity for word cloud creation
entities = [i.lower().replace(' ', '_') for i in entities]
```

```python
# Get geopolitical entities
print([ent.text for ent in doc.ents if ent.label_ == 'GPE'])
```

```python
# List of all the people named in the articles
print([ent.text for ent in doc.ents if ent.label_ == 'PERSON'])
```

```python
# List of all the nationalities and origins named in the articles
print([ent.text for ent in doc.ents if ent.label_ == 'NORP'])
```

```python
# Generate wordcloud
wc = WordCloud().generate(' '.join(entities))
plt.imshow(wc)
```
